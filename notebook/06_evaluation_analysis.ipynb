# Analisis Diversitas Rekomendasi
Mengukur apakah sistem menghasilkan rekomendasi yang:

- Terlalu homogen

- Terlalu terfokus pada satu perusahaan

```python
def diversity_score(recs):
    return recs["companyName"].nunique() / len(recs)

diversity_scores = []

for idx in sample_indices:
    recs = recommend_jobs(idx, top_n=5)
    diversity_scores.append(diversity_score(recs))

pd.Series(diversity_scores).describe()
```
