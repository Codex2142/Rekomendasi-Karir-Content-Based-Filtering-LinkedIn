# Import Lib

```python
import pandas as pd
```

# Read csv

```python
df = pd.read_csv("../dataset/raw/dump.csv")
```

# hapus kolom tidak penting

```python
drop_cols = [
    "memberUrn", "companyUrn", "companyUrl", "positionId",
    "followable", "companyHasLogo", "hasPicture", "genderEstimate"
]

df = df.drop(columns=drop_cols, errors="ignore")
```

# pisahkan kolom

```python
text_cols = ["mbrTitle", "posTitle", "companyName", "posLocation"]
num_cols = [
    "ageEstimate", "connectionsCount", "followersCount",
    "companyFollowerCount", "companyStaffCount",
    "avgMemberPosDuration", "avgCompanyPosDuration"
]

df[text_cols] = df[text_cols].fillna("")
df[num_cols] = df[num_cols].fillna(0)
```

# lowercase

```python
def clean_text(text):
    return str(text).lower().strip()

for col in text_cols:
    df[col] = df[col].apply(clean_text)
```

# export csv

```python
df.to_csv("../dataset/cooked/cleaned_data.csv", index=False)
```
