```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
import pickle
```

```python
df = pd.read_csv('../dataset/cooked/preprocessed.csv')
```

```python
tfidf = TfidfVectorizer(
    stop_words='english',
    max_features=5000,     # batasi fitur agar efisien
    ngram_range=(1,2)      # unigram + bigram (lebih akurat)
)
```

```python
tfidf_matrix = tfidf.fit_transform(df['text_clean'])
tfidf_matrix.shape
```

```python
pickle.dump(tfidf, open('../dataset/cooked/tfidf_vectorizer.pkl', 'wb'))
pickle.dump(tfidf_matrix, open('../dataset/cooked/tfidf_matrix.pkl', 'wb'))
```
